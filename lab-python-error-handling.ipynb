def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))

                    raise ValueError("Invalid quantity! Please enter a non-negative value.")

````python
import time

def initialize_inventory(products):
    try:
        inventory = {}
        for product in products:
            count = int(input(f"Enter the inventory count for {product}: "))
            if count < 0:
                print(f"Invalid input for {product}. Inventory count cannot be negative.")
                raise ValueError("Inventory count cannot be negative.")
            inventory[product] = count
        return inventory
    except ValueError as e:
        print(f"Input Error, please enter a valid number: {e}")
        return {}

def get_customer_orders(products):
    
    try:
        customer_orders = set()
        while True:
            order = input(f"Please enter your Order (available products: {', '.join(products)}): ").strip().lower()
            if order in products:
                customer_orders.add(order)
            else:
                print(f"Sorry, {order} is not in our product list.")
                continue
            more_product = input("Do you want to order more products? (yes/no): ").strip().lower()
            if more_product == "no":
                break
        return customer_orders
    
    except ValueError as e:
        print(f"Input Error, please enter a valid number: {e}")
    except Exception as e:
        print(f"An error occurred while getting customer orders: {e}")
        return set()

def update_inventory(customer_orders, inventory):
    print("Processing Orders...")
    for item in customer_orders:
        if inventory.get(item, 0) > 0:
            inventory[item] -= 1

def calculate_order_statistics(customer_orders, products):
    unique_products_ordered = len(customer_orders)
    percentage_ordered = (unique_products_ordered / len(products)) * 100
    return unique_products_ordered, percentage_ordered

def print_order_statistics(order_status):
    print(f"Order Statistics:-")
    print(f"Total Items Ordered: {order_status[0]}")
    print(f"Percentage of Product Types Ordered: {order_status[1]:.2f}%")

def print_updated_inventory(inventory):
    print("Updated Inventory:-")
    for product, quantity in inventory.items():
        print(f"{product.capitalize()}: {quantity}")

products = ["t-shirt", "mug", "hat", "book", "keychain"]

try:
    inventory = initialize_inventory(products)
    if not inventory:
        print("Inventory initialization failed. Exiting program.")
    else:
        customer_orders = get_customer_orders(products)
        time.sleep(0.5)
        if not customer_orders:
            print("No valid orders entered.")
        else:
            print(f"Summary of items ordered: {customer_orders}")
            time.sleep(1)
            total, percentage = calculate_order_statistics(customer_orders, products)
            time.sleep(1)
            print_order_statistics((total, percentage))
            update_inventory(customer_orders, inventory)
            time.sleep(1)
            print_updated_inventory(inventory)

except Exception as e:
    print(f"An unexpected error occurred: {e}")
````
