````python
import time

def initialize_inventory(products):
    inventory = {}
    for product in products:
        while True:
            try:
                count = int(input(f"Enter the inventory count for {product}: "))
                if count < 0:
                    print("Error: Inventory count cannot be negative. Please try again.")
                    continue
                inventory[product] = count
                break 
            except ValueError:
                print("Error: Please enter a valid whole number for the count.")
    return inventory

def get_customer_orders(inventory):
    customer_orders = []
    while True:
        try:
            num_orders = int(input("\nHow many items would you like to order today? "))
            if num_orders < 0:
                print("Error: You cannot order a negative number of items.")
                continue
            break
        except ValueError:
            print("Error: Please enter a valid number for the quantity of orders.")

    for i in range(num_orders):
        while True:
            order = input(f"Enter item #{i+1} (Available: {', '.join(inventory.keys())}): ").strip().lower()
            
            if order not in inventory:
                print(f"Error: Sorry, '{order}' is not in our product list.")
            elif inventory[order] <= 0:
                print(f"Error: Sorry, '{order}' is currently out of stock!")
            else:
                customer_orders.append(order)
                break
                
    return customer_orders

def update_inventory(customer_orders, inventory):
    print("\nProcessing Orders...")
    for item in customer_orders:
        if inventory.get(item, 0) > 0:
            inventory[item] -= 1
            if inventory[item] == 0:
                print(f"Alert: {item} is now out of stock!")

def calculate_order_statistics(customer_orders, total_products_count):
    """Calculates basic stats about the order."""
    unique_products_ordered = len(set(customer_orders))
    percentage_ordered = (unique_products_ordered / total_products_count) * 100
    return unique_products_ordered, percentage_ordered

def calculate_total_price(customer_orders):
    """Calculates total price with error handling for each item's price."""
    total_price = 0
    unique_orders = set(customer_orders)
    
    prices = {}
    for item in unique_orders:
        while True:
            try:
                price = float(input(f"Enter the price for {item}: "))
                if price < 0:
                    print("Error: Price cannot be negative.")
                    continue
                prices[item] = price
                break
            except ValueError:
                print("Error: Please enter a valid numeric value for the price.")
    
    for item in customer_orders:
        total_price += prices[item]
        
    return total_price

def main():
    products_list = ["t-shirt", "mug", "hat", "book", "keychain"]

    try:
        inventory = initialize_inventory(products_list)
        print(f"\nInitial Inventory: {inventory}")

        customer_orders = get_customer_orders(inventory)
        
        if not customer_orders:
            print("No orders were placed !.")
            return
        total_unique, percentage = calculate_order_statistics(customer_orders, len(products_list))
        print(f"\nOrder Summary: {customer_orders}")
        print(f"Total Unique Products Ordered: {total_unique}")
        print(f"Percentage of Catalog Ordered: {percentage:.2f}%")
        
        time.sleep(2)
        update_inventory(customer_orders, inventory)
        
        total_price = calculate_total_price(customer_orders)
        print(f"\nFinal Transaction Total: ${total_price:.2f}")
        
        print("\nFinal Inventory Status:")
        for product, quantity in inventory.items():
            print(f"- {product.capitalize()}: {quantity}")

    except Exception as e:
        print(f"A critical system error occurred: {e}")

if __name__ == "__main__":
    main()
````
